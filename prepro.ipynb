```python
#컬럼순서바꾸기
#cols = train.columns.tolist()
cols = ['order_id',
        'user_id',
        'product_id',
         'add_to_cart_order',
         'order_number',
         'order_dow',
         'order_hour_of_day',
         'days_since_prior_order',
         'reordered']
```

```python
train = train.rename( index=str, columns={'add_to_cart_order':'add_order', 'order_hour_of_day':'order_hr','days_since_prior_order':'days_since'})
```

```python
prd_count = train.product_id.value_counts(normalize=True)
prd_count
pcnt_rate = pd.DataFrame(train.product_id.value_counts(normalize=True))
```

```python
#제품 별 주문수 
prd_cnt = train.groupby('product_id').count()[['order_id']]
prd_cnt.columns = ['prd_cnt']
train1 = train.merge(prd_cnt, left_on='product_id', right_index=True)
train1.head()
# prd_cnt[['product_id','counts']].groupby(['product_id'],as_index=False).mean().sort_values(by='counts',ascending=False)
```

```python
#가입한지 오래된 고객들
oldUsers = train[['user_id','days_since']].groupby(['user_id'],as_index=False).sum().sort_values(by='days_since',ascending=False)
```

```python
train1 = pd.merge(left=train1, right=oldUsers, how='left', on='user_id')
```
